```python
import sqlite3
import pandas as pd

# Kết nối đến cơ sở dữ liệu (hoặc tạo mới nếu chưa có)
conn = sqlite3.connect('Student.db')
cursor = conn.cursor()

# Tạo bảng Student
cursor.execute('''CREATE TABLE IF NOT EXISTS Student (
                    student_id INT PRIMARY KEY,
                    name VARCHAR(50),
                    class VARCHAR(50),
                    course_id INT,
	                score FLOAT
                )''')
# Chèn dữ liệu
cursor.executemany('INSERT INTO Student (student_id, name, class, course_id, score) VALUES (?, ?, ?, ?,?)',[
    (1, 'Nguyen Minh Hoang', 'May Tinh', 12 , 6.7),
    (2, 'Tran Thi Lan', 'Kinh Te', 34 , 9.2),
    (3, 'Pham Van Nam ','Toan Tin', None , 7.9),
    (4, 'Le Thanh Huyen', 'Toan Tin', 20 , 7.2),
    (5, 'Vu Quoc Anh', 'May Tinh', 24 , 8.0),
    (6, 'Dang Thuy Linh', 'May Tinh', 24 , 5.5),
    (7, 'Bui Tien Dung', 'Kinh Te', 34 , 9.2),
    (8, 'Ho Ngoc Mai', 'Toan Tin', 20 , 8.8),
    (9, 'Duong Huu Phuc', 'Kinh Te', None , 7.2),
    (10, 'Cao Thi Hanh', 'May Tinh', None , 7.0)
 
])

# Tạo bảng Course
cursor.execute('''CREATE TABLE IF NOT EXISTS Course (
                    id INT PRIMARY KEY,
                    course_name VARCHAR(50)
    )
''')
# Chèn dữ liệu 
cursor.executemany('INSERT INTO Course (id, course_name) VALUES (?, ?)', [
    (12, 'Giai Tich'),
    (34, 'Thong Ke'),
    (26, 'Tin Hoc')
])
```

```python
# Lấy toàn bộ thông tin
cursor.execute("SELECT * FROM Student")
df = pd.DataFrame(cursor.fetchall(), columns=['student_id', 'name', 'class', 'course_id','score'])
print("Toàn bộ thông tin:")
display(df)
```

```python
# Decartes
query = '''
    SELECT * FROM Student, Course;
'''
df = pd.read_sql(query, conn)
df
```

```python
# INNER JOIN
query = '''
    SELECT Student.*, Course.course_name
    FROM Student
    INNER JOIN Course ON Student.course_id = Course.id;
'''
df = pd.read_sql(query, conn)
df
```

```python
# LEFT JOIN
query = '''
    SELECT Student.*, Course.course_name
    FROM Student
    LEFT JOIN Course ON Student.course_id = Course.id;
'''
df = pd.read_sql(query, conn)
df
```

```python
# RIGHT JOIN
query = '''
    SELECT Course.*, Student.name, Student.class, Student.score
    FROM Course
    LEFT JOIN Student ON Student.course_id = Course.id;
'''
df = pd.read_sql(query, conn)
df
```

```python
# FULL OUTER JOIN
query = '''
    SELECT Student.*, Course.course_name
    FROM Student
    LEFT JOIN Course ON Student.course_id = Course.id
    UNION
    SELECT Student.*, Course.course_name
    FROM Student
    RIGHT JOIN Course ON Student.course_id = Course.id;
'''
df = pd.read_sql(query, conn)
df
```

```python

# Cập nhật course_id còn thiếu bằng một giá trị hợp lệ từ bảng Course
cursor.execute('''UPDATE Student 
    SET course_id = (SELECT id FROM Course ORDER BY RANDOM() LIMIT 1) 
    WHERE course_id IS NULL
''')

# Xóa các sinh viên có course_id không tồn tại trong Course
cursor.execute('''
    DELETE FROM Student 
    WHERE course_id NOT IN (SELECT id FROM Course)
''')
```

```python
query_a = '''
    SELECT class, COUNT(student_id) AS total_students, AVG(score) AS avg_score 
    FROM Student 
    GROUP BY class
'''
df_a = pd.read_sql_query(query_a, conn)
print("Tổng số sinh viên và điểm trung bình của từng lớp:")
print(df_a)
```

```python
query_b = '''
    SELECT Course.course_name, COUNT(Student.student_id) AS total_students, AVG(Student.score) AS avg_score
    FROM Student
    JOIN Course ON Student.course_id = Course.id
    GROUP BY Course.course_name
'''
df_b = pd.read_sql_query(query_b, conn)
print("Tổng số sinh viên và điểm trung bình của từng môn học:")
print(df_b)
```

```python
query_c = '''
    SELECT Course.course_name, 
           AVG(Student.score) AS avg_score,
           CASE 
               WHEN AVG(Student.score) >= 9.0 THEN 'Xuất sắc'
               WHEN AVG(Student.score) BETWEEN 6.0 AND 8.9 THEN 'Tốt'
               ELSE 'Kém'
           END AS classification
    FROM Student
    JOIN Course ON Student.course_id = Course.id
    GROUP BY Course.course_name
'''
df_c = pd.read_sql_query(query_c, conn)
print("Phân loại thi đua theo số điểm của từng môn học:")
print(df_c)
```

```python
query_top_highest = '''
    SELECT student_id, name, class, score
    FROM Student
    ORDER BY score DESC
    LIMIT 3
'''
df_top_highest = pd.read_sql_query(query_top_highest, conn)
print("Top 3 sinh viên có điểm cao nhất:")
print(df_top_highest)

query_top_lowest = '''
    SELECT student_id, name, class, score
    FROM Student
    ORDER BY score ASC
    LIMIT 3
'''
df_top_lowest = pd.read_sql_query(query_top_lowest, conn)
print("Top 3 sinh viên có điểm thấp nhất:")
print(df_top_lowest)
```

```python
query_class_highest = '''
    SELECT student_id, name, class, score
    FROM (
        SELECT student_id, name, class, score,
               RANK() OVER (PARTITION BY class ORDER BY score DESC) AS rank_high
        FROM Student
    ) WHERE rank_high <= 3
'''
df_class_highest = pd.read_sql_query(query_class_highest, conn)
print("Top 3 sinh viên đạt thứ hạng cao nhất theo từng lớp:")
print(df_class_highest)

query_class_lowest = '''
    SELECT student_id, name, class, score
    FROM (
        SELECT student_id, name, class, score,
               RANK() OVER (PARTITION BY class ORDER BY score ASC) AS rank_low
        FROM Student
    ) WHERE rank_low <= 3
'''
df_class_lowest = pd.read_sql_query(query_class_lowest, conn)
print("Top 3 sinh viên đạt thứ hạng thấp nhất theo từng lớp:")
print(df_class_lowest)
```

```python
query_course_highest = '''
    SELECT student_id, name, course_id, score
    FROM (
        SELECT student_id, name, course_id, score,
               RANK() OVER (PARTITION BY course_id ORDER BY score DESC) AS rank_high
        FROM Student
    ) WHERE rank_high <= 3
'''
df_course_highest = pd.read_sql_query(query_course_highest, conn)
print("Top 3 sinh viên đạt thứ hạng cao nhất theo từng môn học:")
print(df_course_highest)

query_course_lowest = '''
    SELECT student_id, name, course_id, score
    FROM (
        SELECT student_id, name, course_id, score,
               RANK() OVER (PARTITION BY course_id ORDER BY score ASC) AS rank_low
        FROM Student
    ) WHERE rank_low <= 3
'''
df_course_lowest = pd.read_sql_query(query_course_lowest, conn)
print("Top 3 sinh viên đạt thứ hạng thấp nhất theo từng môn học:")
print(df_course_lowest)
```

```python
from datetime import datetime, timedelta
# Thêm cột graduation_date vào bảng Student

cursor.execute("ALTER TABLE Student ADD COLUMN graduation_date DATETIME")

# Đọc dữ liệu vào Pandas DataFrame
df = pd.read_sql("SELECT * FROM Student", conn)

# Tính toán ngày tốt nghiệp: ngày hiện tại + (10 - score) tháng
df['graduation_date'] = df['score'].apply(lambda x: (datetime.now() + timedelta(days=(10 - x) * 30)).strftime("%Y-%m-%d %H:%M:%S") if pd.notna(x) else None)

# Lưu dữ liệu đã cập nhật vào SQLite
df.to_sql("Student", conn, if_exists="replace", index=False)
df
```
